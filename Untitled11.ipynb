```python
import os
import urllib.request
import zipfile
import time
import sys
import math
import hashlib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('seaborn-v0_8-whitegrid')
plt.rc('font', family='Malgun Gothic')
plt.rc('axes', unicode_minus=False)

url = "https://archive.ics.uci.edu/static/public/352/online+retail.zip"
zip_path = "online_retail.zip"
csv_path = "Online Retail.csv"

if not os.path.exists(csv_path):
    urllib.request.urlretrieve(url, zip_path)
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(".")

class BloomFilter:
    def __init__(self, size, num_hashes):
        self.size = size
        self.num_hashes = num_hashes
        self.bit_array = np.zeros(size, dtype=bool)

    def _hashes(self, item):
        h1 = int(hashlib.md5(str(item).encode()).hexdigest(), 16)
        h2 = int(hashlib.sha256(str(item).encode()).hexdigest(), 16)
        for i in range(self.num_hashes):
            yield (h1 + i * h2) % self.size

    def add(self, item):
        for bit_index in self._hashes(item):
            self.bit_array[bit_index] = True

    def check(self, item):
        for bit_index in self._hashes(item):
            if not self.bit_array[bit_index]:
                return False
        return True

class CountMinSketch:
    def __init__(self, width, depth):
        self.width = width
        self.depth = depth
        self.table = np.zeros((depth, width), dtype=int)

    def _hash(self, item, depth_index):
        h = int(hashlib.sha256(f"{item}_{depth_index}".encode()).hexdigest(), 16)
        return h % self.width

    def add(self, item, count=1):
        for d in range(self.depth):
            w = self._hash(item, d)
            self.table[d, w] += count

    def estimate(self, item):
        res = sys.maxsize
        for d in range(self.depth):
            w = self._hash(item, d)
            res = min(res, self.table[d, w])
        return res

df = pd.read_csv(csv_path, encoding="ISO-8859-1")
stream_data = df['StockCode'].dropna().astype(str).tolist()

exact_set = set(stream_data)
exact_counter = {}
for item in stream_data:
    exact_counter[item] = exact_counter.get(item, 0) + 1

non_existing_items = [f"NON_EXIST_{i}" for i in range(10000)]

bf_sizes = [100000, 500000, 1000000]
bf_results = []

for size in bf_sizes:
    t_start = time.time()
    bf = BloomFilter(size=size, num_hashes=4)
    for item in stream_data:
        bf.add(item)
    t_end = time.time()

    fp_count = sum(1 for item in non_existing_items if bf.check(item))
    fpr = fp_count / len(non_existing_items)
    mem_size = bf.bit_array.nbytes / 1024

    bf_results.append({
        'Size': size, 'FPR': fpr, 'Memory(KB)': mem_size, 'Time(s)': t_end - t_start
    })

cms_widths = [1000, 5000, 10000]
cms_results = []

for width in cms_widths:
    t_start = time.time()
    cms = CountMinSketch(width=width, depth=5)
    for item in stream_data:
        cms.add(item)
    t_end = time.time()

    errors = []
    for item in exact_counter.keys():
        est = cms.estimate(item)
        errors.append(abs(est - exact_counter[item]))
    mae = np.mean(errors)
    mem_size = cms.table.nbytes / 1024

    cms_results.append({
        'Width': width, 'MAE': mae, 'Memory(KB)': mem_size, 'Time(s)': t_end - t_start
    })

df_bf = pd.DataFrame(bf_results)
df_cms = pd.DataFrame(cms_results)

print("\n[Bloom Filter 실험 결과]")
print(df_bf.to_string(index=False))

print("\n[Count-Min Sketch 실험 결과]")
print(df_cms.to_string(index=False))

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

sns.lineplot(data=df_bf, x='Memory(KB)', y='FPR', marker='o', ax=axes[0], color='blue')
axes[0].set_title("Bloom Filter: Memory Size vs False Blue Positive Rate")
axes[0].set_ylabel("FPR")

sns.lineplot(data=df_cms, x='Memory(KB)', y='MAE', marker='s', ax=axes[1], color='red')
axes[1].set_title("Count-Min Sketch: Memory Size vs MAE")
axes[1].set_ylabel("MAE")

plt.tight_layout()
plt.savefig('streaming_tradeoff_report.png', dpi=300)
plt.show()
```
